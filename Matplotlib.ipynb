```python
df = pd.read_csv('company_sales_data.csv')
df.head(4)
```

```python
ypoints= df['total_profit'].tolist()
xpoints= df['month_number'].tolist()
plt.plot(xpoints, ypoints)
plt.title("Total Income")
plt.xlabel("Month number")
plt.ylabel("Total Profit")
plt.show()
```

```python
ypoints= df['total_profit'].tolist()
xpoints= df['month_number'].tolist()
plt.plot(xpoints, ypoints, label = "Profit data of Last Year" ,  lw = 3, mfc = "k", color = 'red', ls = 'dotted', marker ='o')
plt.title("Total Income")
plt.xlabel("Month number")
plt.ylabel("Total Profit")
plt.legend(loc='lower right')
plt.show()
```

```python
xpoints = df['month_number'].to_list()
facecream = df['facecream'].to_list()
facewash = df['facewash'].to_list()
toothpaste = df['toothpaste'].to_list()
bathingsoap = df['bathingsoap'].to_list()
plt.plot(xpoints, facecream , label = 'Face Cream Sales Data', c="r", marker = 'o')
plt.plot(xpoints, facewash , label = 'Face Wash Sales Data', c= 'b', marker = 'o')
plt.plot(xpoints, toothpaste , label = 'Tooth Paste Sales Data', c = 'black', marker = 'o')
plt.plot(xpoints, bathingsoap , label = 'Bathingsoap Sales Data', c= 'pink', marker = 'o')
plt.legend(loc='upper left')
plt.ylabel('Sales Unit In Number')
plt.xlabel('Month Number')
plt.title("Sales Data")
plt.show()
```

```python
ypoints = df['toothpaste'].to_list()
xpoints = df['month_number'].to_list()
plt.scatter(xpoints, ypoints, label = "Tooth Paste Sales Data")
plt.grid(ls = 'dotted')
plt.xlabel("Month Number")
plt.ylabel("Number Of Units Sold")
plt.title("Tooth Paste Sales Data")
plt.legend()
plt.show()
```

```python
xpoints = df['month_number'].to_list()
facewash = df['facewash'].to_list()
facecream = df['facecream'].to_list()
plt.bar([a-0.2 for a in xpoints], facewash,   align = 'edge',label = 'Facewash Sales Data',width = 0.25)
plt.bar([a+0.2 for a in xpoints],facecream,   align = 'edge' , label = 'Face Cream Sales Data', width = -0.25)
plt.legend(loc='upper left')
plt.show()
```

```python
xpoints = df['month_number'].to_list()
ypoints = df['bathingsoap'].to_list()
plt.grid(ls= 'dotted')
plt.xlabel("Month Number")
plt.ylabel("Bathing Soap Sales")
plt.title("Bathing Soap Sales Data")
plt.bar(xpoints, ypoints)
plt.savefig("bathingsoap4.png", dpi = 300, bbox_inches = 'tight')
plt.show()
```
