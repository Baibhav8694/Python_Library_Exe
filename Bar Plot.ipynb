# Use of the keywords In Histplot to Use it and make awesome Graphs for data visualisation 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset("tips")
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, label = "Total Bill", axlabel = "Total Bill Of Restraunt")
plt.legend()
sns.set()
plt.title("Histogram of Total BIll")
```

# How to style the plot and Other Stuff 

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], kde = False, bins = bins)
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
bins = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], kde = True, bins = bins, hist_kws = {'color':'red', 'edgecolor': 'blue',
                                                                         'lw': 2, 'ls': '--', 'alpha': 0.9}
            , kde_kws = {'color':'yellow',
                                         'lw': 5, 'ls': 'solid', 'alpha': 0.9})
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, 
             label = "Total Bill", axlabel = "Total Bill Of Restraunt",
             hist_kws = {'color': 'red','edgecolor':'blue', 'ls': '--', 'lw': 3, 'alpha':0.8}
            , fit_kws = {"color": 'white', 'lw': 2, 'ls': '--' })
plt.legend()
sns.set()
plt.title("Histogram of Total BIll")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.distplot(tips_df["total_bill"], bins = 5 , label='total bill',
            hist_kws = {'density': 20})
sns.distplot(tips_df['tip'], bins = 2 , label = 'tip')
sns.distplot(tips_df["size"], bins = 2 , label = 'size')
plt.legend()
plt.show()
```
