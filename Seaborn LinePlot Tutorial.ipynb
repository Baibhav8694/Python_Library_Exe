```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head(5)
```

```python
sns.lineplot(x = "tip", y = "total_bill", data = tips_df)
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df)
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "sex")
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker")
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker")
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", palette= 'hot')
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", palette= 'hot', dashes = False)
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", palette= 'hot', markers = ['o', 'p'])
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", 
             palette= 'hot', legend = "brief")
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", 
             palette= 'hot', markers = ['o', 'p'])
```

```python
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", 
             palette= 'hot', markers = ['o', 'p'])
plt.xlabel("Size", fontsize = 20, color = "red")
plt.ylabel("Total Bill", fontsize = 20, color = "blue")
plt.show()
```

```python
sns.set(style = "darkgrid")
sns.lineplot(x = 'size', y = "total_bill", data = tips_df, hue = "smoker", style = "smoker", 
             palette= 'hot', markers = ['o', 'p'])
plt.xlabel("Size", fontsize = 20, color = "red")
plt.ylabel("Total Bill", fontsize = 20, color = "blue")
plt.legend(loc = "upper left")
plt.savefig("lineplot.png", dpi = 100)
plt.show()
```
