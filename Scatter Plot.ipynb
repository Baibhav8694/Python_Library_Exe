# Scatterr Plot Tutorial

To know about scatter plot go through the following the further tutorials

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic_df = sns.load_dataset("titanic")
```

```python
titanic_df.head(5)
```

```python
titanic_df.shape
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue= "sex")
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue= "sex", style = 'who', size = "who", sizes = (100, 400))
```

```python
plt.figure(figsize=(16, 9))
sns.scatterplot(x = 'who', y = 'fare', data = titanic_df, hue= "alive", style = 'alive', size = 'who', sizes = (100, 300)
               ,palette = 'inferno_r', alpha = 0.7)
```
