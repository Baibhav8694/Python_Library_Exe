## Seaborn HistPlot Tutorial 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.head(5)
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.histplot(tips_df["size"])
```

```python
sns.distplot(tips_df["total_bill"], bins = 20)
```

```python
sns.distplot(tips_df["total_bill"], bins = 50)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, hist = False)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, hist = True, kde = False)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, rug = True)
```

```python
from scipy.stats import norm
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, color = "red")
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, vertical = True)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, norm_hist = True)
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, axlabel = "Total Bill Of Restraunt")
```

```python
sns.distplot(tips_df["total_bill"], bins = 20, kde = False, fit = norm, label = "Total Bill", axlabel = "Total Bill Of Restraunt")
plt.legend()
plt.title("Histogram of Total BIll")
```
